```python
#Q1: Create a dataframe which contains information of a company's products:
data=pd.DataFrame({'prod_nam':['g1','g2','c1','c2','c3','f1','f2'],
                 'price':[45,25,75,35,89,20,55],
                 'monthly_demand':[1200,2500,200,1850,120,3500,545],
                 'brand':['z','x','x','y','z','y','z']})
data
```

```python
#Q1-a: Add a column to calculate monthly revenue of each product
data['monthly_revenue']=data['price']*data['monthly_demand']
data
```

```python
#Q1-b: Calculate total monthly revenue generated by brand z
np.sum(data.loc[data['brand']=='z',['prod_nam','price','monthly_revenue']])
```

```python
np.sum(data.loc[(data['brand']=='z'),'monthly_revenue'])  
```

```python
#Q1-c: Create a column called revenue share which includes monthly revenue share of each product
data['revenue_share']=data['monthly_revenue']*0.5
data
```

```python
#Q1-d way 1: Which brand does have the largest revenue share?
#np.max(data['revenue_share'])
data[data['revenue_share']==data['revenue_share'].min()]
```
